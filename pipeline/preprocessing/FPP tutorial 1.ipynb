```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
os.chdir("C:\\Users\\Oke\\Downloads")
tute1 = pd.read_csv("tute1.csv")
tute1.head()
```

```python
# Convert to quarterly period and set as index
tute1['Quarter'] = pd.to_datetime(tute1['Quarter']).dt.to_period('Q')
tute1.set_index('Quarter', inplace=True)

# Convert Period index to timestamp for plotting
quarter_timestamps = tute1.index.to_timestamp()

# Then plot with smoother styling
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

axes[0].plot(quarter_timestamps, tute1['Sales'], linewidth=2)
axes[0].set_title('Sales vs Quarter')
axes[0].set_ylabel('Sales')
axes[0].grid(True, alpha=0.3)

axes[1].plot(quarter_timestamps, tute1['AdBudget'], linewidth=2, color='orange')
axes[1].set_title('AdBudget vs Quarter')
axes[1].set_ylabel('AdBudget')
axes[1].grid(True, alpha=0.3)

axes[2].plot(quarter_timestamps, tute1['GDP'], linewidth=2, color='green')
axes[2].set_title('GDP vs Quarter')
axes[2].set_ylabel('GDP')
axes[2].set_xlabel('Quarter')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
